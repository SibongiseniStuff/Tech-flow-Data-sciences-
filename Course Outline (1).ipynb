## Task 3: Product Sales Analysis

### Extract the two columns of data: "Sales" and "Sale Amount". Group by the sales's name and calculate the performance (Sale amount Sum) of each salesperson.

```python
yewu_data=odata[['Sales','Sale Amount']].groupby(by='Sales').sum()
```

```python
%time
yewu_data
```

### Extract the two columns of data: "Month" and "Sale amount Sum". Group by month, calculate the total sales amount for each month of the year, and save the result to show_1.

```python
show_1=odata[['Month','Sale amount Sum']].groupby(by='Month').sum()
show_1
```

### Extract the two columns of data: "Quarter" and "Sale amount Sum". Group by quarter and calculate the total sales amount for each quarter of the year

```python
jidu_data=odata[['Quarter','Sale amount Sum']].groupby(by='Quarter').sum()
jidu_data
```

### Extract the three columns of data: "Product", "Quarter" and "Sale Amount". Group by product and quarter, and calculate the total sales volume of each product in each quarter of the year.

```python
cp_jd_data=odata[['Product','Quarter','Sale amount Sum']].groupby(by=['Product','Quarter']).sum()
cp_jd_data
```

## Task 4: Customer Purchase Demand Analysis

### Extract the three columns: "Customer", "Product", and "Sale Amount". Group by customer and product, calculate the total sales volume (demand) of the product Door Bolt for each customer in the second half of the year, and save the result to show_2.

```python
show_2=odata.loc[(odata.Month>6) & (odata.Product=='Door Bolt')][['Customer','Sale Amount']].groupby(by=['Customer']).sum()
show_2
```

### Use a pivot table to calculate the total purchase amount for each customer in each quarter throughout the year.
Display the grand total and name the summary field "Total Sales Amount".

```python
pd.pivot_table(odata,values='Sale amount Sum',columns='Quarter',index='Customer', aggfunc='sum', margins=True, margins_name='Sale amount Sum')
```

### Use a pivot table to calculate the total purchase quantity of each product category by each customer in each quarter of the year, and replace missing values with 0.

```python
pd.pivot_table(odata, index=['Quarter','Product'],values='Sale Amount', columns='Customer',aggfunc='sum',fill_value=0).reset_index()
```

### Use the data results from show_1 to display the total monthly sales amount across the year with a line chart.
(Values are presented in ten thousands and rounded to two decimal places).

```python
show_1=odata[['Month','Sale amount Sum']].groupby(by='Month').sum()
show_1
```

```python
show_1['Sale amount Sum'] = round(show_1['Sale amount Sum']/10000,2)
```

```python
show_1
```

```python
plt.figure(figsize=(12,7),facecolor='Lightgreen')
plt.title('Year Sales', fontsize=24, color='red')

plt.xlabel('Month',fontsize=12, color='red')
plt.ylabel('Sale amount Sum',fontsize=12, color='red')

plt.plot(show_1.reset_index()['Month'],show_1['Sale amount Sum'],label='Sale Amount',
         linestyle='-.',linewidth=2.5,color='red', marker='o',markersize=7,alpha=0.8)

for x,y in enumerate(show_1['Sale amount Sum']):
        plt.text(x+1, y+5,'%.2f' %y, color='blue', fontsize=13)

plt.legend(fontsize=12)
plt.show()
```

### Use the results from show_2 to create a bar chart showing customers' demand for door bolts in the first half of the year.
Sort the bars in descending order from left to right by demand volume.
（Present the values in ten thousands and keep two decimal places).

```python
plt.figure(figsize=(12,7),facecolor='Lightgreen')
plt.title('Year Sales', fontsize=24, color='red')

plt.xlabel('Month',fontsize=12, color='red')
plt.ylabel('Sale amount Sum',fontsize=12, color='red')

months=show_1.reset_index()['Month']
values=show_1['Sale amount Sum']

plt.bar(months,values,label='Sale Amount',color='red',edgecolor='yellow',alpha=0.8)
for x,y in enumerate(show_1['Sale amount Sum']):
        plt.text(x+1, y+5,'%.2f' %y, color='blue', fontsize=13)

plt.legend(fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8), facecolor='lightgreen')

plt.title('Year Sales Distribution by Month', fontsize=20, color='red')

# Extract data
months = show_1.index
values = show_1['Sale amount Sum']

# Pie chart
plt.pie(
    values,
    labels=months,
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'edgecolor': 'black'}
)

month_names = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

plt.figure(figsize=(10,8), facecolor='lightgreen')

plt.pie(
    show_1['Sale amount Sum'],
    labels=month_names,
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Year Sales Distribution by Month', fontsize=20, color='red')
plt.tight_layout()
plt.show()
plt.tight_layout()
plt.show()
```
